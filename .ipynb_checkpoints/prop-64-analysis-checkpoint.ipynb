### Importing a list of committees & saving into a variable

### Importing a list of contributions

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

### Counting the number of contributions under each committee

```python
contribs.committee_name.value_counts()
```

```python
contribs[contribs.calaccess_committee_id==1371855].info()
```

```python
contribs.head()
```

### Saving props on Prop 64 into variable "prop"

```python
prop = props[props.prop_name=="PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE."]
```

```python
prop.head()
```

### Because both "prop" and "contribs" have the same column "calaccess_committee_id", we use the merge method to return all "contribs" that are related to Prop 64

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

### Comparing number of contributions on each side on Prop 64

```python
merged.committee_position.value_counts().reset_index()
```

```python
merged.amount.sum()
```

### Separating by side

```python
support=merged[merged.committee_position=="SUPPORT"]
```

```python
oppose=merged[merged.committee_position=="OPPOSE"]
```

### Sorting contribution amount in descending order for the opposing side

```python
merged[merged.committee_position=="OPPOSE"].sort_values("amount", ascending=False)
```

### Creating a table showing contribution amount in descending order on the supporting side. Columns "amount" and "committee_name_x" are selected in the table.

```python
support[["amount", "committee_name_x"]].sort_values("amount", ascending=False).head(20)
```

### Combining the first name and last name into full name

```python
top_supporters = merged.groupby(["contributor_firstname", "contributor_lastname"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
top_supporters["contributor_fullname"] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname
```

### Creating a graph showing amount of contribution for each person

```python
top_supporters.amount.plot.barh().set_yticklabels(top_supporters.contributor_fullname)
```

```python
top_supporters.to_csv("top_supporters.csv")
```

### Calculating percentage of contribution from CA in the US

```python
merged[merged.contributor_state=="CA"].amount.sum() / merged.amount.sum()
```

### In the previous graph, the top two contributor are the same person but with variations on their names. The following lines combine two variations into one

```python
top_supporters = merged.groupby(["contributor_firstname", "contributor_lastname"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(11)
```

```python
top_supporters["contributor_fullname"] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname
```

```python
def combine_names(row):
    if row.contributor_fullname.startswith("SEAN PARKER"):
        return "SEAN PARKER"
    return row.contributor_fullname
```

```python
top_supporters["contributor_cleanname"]=top_supporters.apply(combine_names, axis=1)
```

### Updating the list and make another graph with cleaned up names

```python
top_supporters = top_supporters.groupby("contributor_cleanname").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
top_supporters
```

```python
top_supporters.plot.barh().set_yticklabels(top_supporters.contributor_cleanname)
plt.xlabel("amount")
plt.ylabel("contributor name")
plt.title("Contribution Amount")
```
